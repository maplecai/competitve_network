```python
import primer3
import numpy as np
from tqdm import tqdm
from utils import *
```

```python
class PrimerGenerator:
    
    def __init__(self):
        self.nt = ['A','G','C','T']
    
    def generate(self, length):
        seq = ''
        for i in range(length):
            j = np.random.randint(0,4)
            seq += self.nt[j]

        return seq
```

```python
def calcPoly(seq):
    repeat = 0
    max_repeat = 0
    for i in range(1, len(seq)):
        if (seq[i] == seq[i-1]):
            repeat += 1
            max_repeat = max(max_repeat, repeat)
        else:
            repeat = 0
    return max_repeat
```

```python
def calcGC(seq):
    GC = seq.count('C') + seq.count('G')
    GC_content = GC / len(seq)
    return GC_content
```

```python
generator = PrimerGenerator()
generator.generate(20)
```

```python
tm_list = []

for i in tqdm(range(100000)):
    seq = generator.generate(20)
    tm = primer3.calcTm(seq)
    tm_list.append(tm)

print(np.mean(tm_list))
```
